### Create a program to check if a number is divisible by 7 and 11 both

```python
num = int(input('enter a number'))
if num % 7 == 0 and num % 11 == 0:
    print(f'yes {num} divisible by both 7 and 11')
else:
    print(f'number {num} is not divisible by both 7 and 11')
```

```python
num = int(input('enter a number'))
if num % 7 == 0 and num % 11 == 0:
    print(f'yes {num} divisible by both 7 and 11')
elif num % 7 == 0 and not num % 11 == 0:
    print(f'yes {num} divisible by 7')
elif not num % 7 == 0 and num % 11 == 0:
    print(f'yes {num} divisible by 11')
else:
    print(f'number {num} is not divisible by both 7 and 11')
```

>>> WAP to find all prime numbers in range of 100 -500

```python
lower = int(input("Enter lower range: "))
upper = int(input("Enter upper range: "))

for num in range(lower,upper + 1):
    if num > 1:
        for i in range(2,num):
            if (num % i) == 0:
                break
            else:
                print(num)
```
